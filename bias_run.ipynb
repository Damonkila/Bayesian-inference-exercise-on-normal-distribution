## From the color plot of rescaled log_probi we can see that max. log_probi is far away from true $(\mu,\sigma)$. And no local max. near true $(\mu,\sigma)$

## This result is 'accidental', actually near [-39, 4.8], log llh is extremely small that e^llh = 0, but in log scale we have larger valid numerical range so with a extremely small erf() value of selection bias as the denominaor of the log_probi function, log_p[-39,4.8] becomes the max. 
## One approach is to reject the extremely small bias value and assign -inf to log_p 
## Or simply narrow the range of parameter space
